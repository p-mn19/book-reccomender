```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
books = pd.read_csv(f"{path}/books.csv")
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()  # Create a set of axes for the plot (like a blank canvas)
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)
# Create a heatmap showing missing values in the 'books' dataset.
# 'books.isna()' checks for missing values (NaN).
# '.transpose()' flips the rows and columns for better visualization.
# 'cbar=False' removes the color bar.

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(),1,0) 
# create a new column "missing_description" and puts 1 where the description is missing else put 0
books["age_of_book"] = 2025-books["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman") # spearman is used for 
# measuring monotonic relationships(eg: when one variable increases the other also increases)
#it is mostly used with non-linear data

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix,annot=True, fmt=".2f",cmap="coolwarm", 
                      cbar_kws={"label":"Spearman correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

 values range between +1 to -1 
 
 A positive value means variables move in the same direction.
 
A negative value means they move in opposite directions.

```python
books[(books["description"].isna()) |
    (books["num_pages"].isna()) |
    (books["average_rating"].isna()) |
    (books["published_year"].isna()) 
]   
#fetch the books which have any of these columns missing
```

```python
book_missing = books[~(books["description"].isna()) &
    ~(books["num_pages"].isna()) &
    ~(books["average_rating"].isna()) &
    ~(books["published_year"].isna()) 
] 
book_missing
#fetch books that have complete values in all these columns (the opposite of prev code block)
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
book_missing = book_missing.copy()  # Create explicit copy first
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
#seperating books whose description is greater than or equal to 25 words
book_missing_25_words = book_missing[book_missing["words_in_description"]>=25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words = book_missing_25_words.copy()

book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join,axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
(
    book_missing_25_words
    .drop(["subtitle","missing_description","age_of_book","words_in_description"],axis=1)
    .to_csv("books_cleaned.csv",index=False)
)
```
